## Simple Linear Regression

## Import Libraries

```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Load The Data

```python
# load the data
df = sns.load_dataset('tips')
df.head()
```

## Regression Plot

```python
# Create a Regression plot from using seaborn
sns.lmplot(x = 'total_bill', y = 'tip', data = df)
```

## Select the Feature and Target Column

```python
# Select the feature column from df
X = df[['total_bill']]
# Select the target column
y = df['tip']
```

## Call the Linear Model and Train the Data

```python
# Load / Call the Linear Regression Model
model = LinearRegression()
# Fit the model
model.fit(X, y)
```

## Predict the Values

```python
# Prediction the tip for an unknown total_bill
# model.predict([[35.527]])
model.predict([[20.456]])
```
