## Multiple Linear Regression / Multi Linear Regression of Tips

##### Multiple / Multi Linear Regression means we select the multiple featues of tips to predict the target column.

## Import Libraries

##### There are many libraries like numpy, pandas to deal with numeric values. seaborn and matplotlib used for show the data in plots. Sciket Learn which are used for implement Machine Learning Algorithms.

```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Load The Data

##### This is the data of the Tips.

```python
df = sns.load_dataset('tips')
df.head()
```

## Create a Regression Plot

##### Here we make the regression plot by using seaborn (sns.lmplot) to show our data in a linear way.

```python
sns.lmplot(x = 'size', y = 'tip', data = df)
```

## Select the Feature and Target Column

##### Now we select the multiple feature columns and one target variable for prediction in a tips data.

```python
# Select the Feature Column
X = df[['total_bill', 'size']]

# Select the Target Column
y = df['tip']
```

## Load the Model and Train the Data

##### First of all we load the Linear Regression Model because our data is numerical. And after loading the model, we train the model by giving our Features (X) and Target(y) values.

```python
# Load the Linear Regression Model
model = LinearRegression()

# Fit the model
model.fit(X, y)
```

## Predict the Tip

##### Now we predict the tip with the help of model.

```python
# Lets predict the tip 
model.predict([[20.23, 2]])
```
