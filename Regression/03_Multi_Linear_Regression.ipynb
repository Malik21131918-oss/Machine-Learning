```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the Data
df = sns.load_dataset('tips')
df.head()
```

```python
# Encode Smoker Column
df['smoker_num'] = df['smoker'].map({'Yes' : 1, 'No' : 0})
df.head()
```

```python
# Create a regression plot by using Seaborn
sns.lmplot(x = 'total_bill', y = 'tip', data = df)
```

```python
# Select the featue column
X = df[['total_bill', 'size', 'smoker_num']]

# Select the target column
y = df['tip']
```

```python
# Load/ Call Linear Regression Model
model = LinearRegression()

# Fit the model / Train the model
model.fit(X, y)
```

```python
# Predict the tip
print(model.predict([[35.527, 5, 1]]))
print(model.predict([[35.527, 5, 0]]))
```
